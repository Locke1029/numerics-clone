# Condition number of a matrix
and how it relates to the accuracy of solving a linear system

```python
import numpy as np
```

```python
n = 100

U = np.linalg.qr(np.random.randn(n, n))[0]
V = np.linalg.qr(np.random.randn(n, n))[0]
```

```python
s = np.logspace(-8, 0, n)
A = U @ np.diag(s) @ V.T

s = np.logspace(-2, 0, n)
B = U @ np.diag(s) @ V.T

print(f'Condition number of A is {np.linalg.cond(A):.2e}')
print(f'Condition number of B is {np.linalg.cond(B):.2e}')
```

```python
x = np.ones(n)

xhat = np.linalg.solve(A, A@x)
print(f'Error solving A is {np.linalg.norm(xhat - x):.2e}')

xhat = np.linalg.solve(B, B@x)
print(f'Error solving B is {np.linalg.norm(xhat - x):.2e}')
```
